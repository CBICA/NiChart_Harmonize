### View trends of harmonized data in test1

```python
import pandas as pd
import numpy as np
import sys
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Load data
dfA = pd.read_csv('../Out/test1_df_out_A.csv')
```

```python
ROI = 'MUSE_Volume_601'
sns.lmplot(data = dfA, x = 'Age', y = ROI, hue = 'SITE', col = 'IsF')
plt.show()
```
